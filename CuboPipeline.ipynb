```python
# Definio dos pontos que definem os vrtices do cubo
P1 = (-0.5, -0.5, -0.5)
P2 = (-0.5, -0.5, 0.5)
P3 = (-0.5, 0.5, -0.5)
P4 = (-0.5, 0.5, 0.5)
P5 = (0.5, -0.5, -0.5)
P6 = (0.5, -0.5, 0.5)
P7 = (0.5, 0.5, -0.5)
P8 = (0.5, 0.5, 0.5)
#b) Definio das arestas do cubo
# Definio das arestas do cubo
arestas = [(P1, P2), (P2, P4), (P4, P3), (P3, P1),
(P5, P6), (P6, P8), (P8, P7), (P7, P5),
(P1, P5), (P2, P6), (P3, P7), (P4, P8)]


import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Plotando o cubo
fig = plt.figure() # Criando uma figura
ax = fig.add_subplot(projection='3d') # Adicionando um subplot tridimensional  figura
# Plotando as arestas do cubo
for aresta in arestas:
  ponto1 = aresta[0] # Obtendo as coordenadas do primeiro ponto da aresta (x,y,z)
  ponto2 = aresta[1] # Obtendo as coordenadas do segundo ponto da aresta (x,y,z)
# Plotando uma linha entre os dois pontos para representar a aresta
  ax.plot([ponto1[0], ponto2[0]], [ponto1[1], ponto2[1]], [ponto1[2], ponto2[2]], 'b')
#Coordenadas 3D (x,y,z)
#1) Modelagem do objeto
# Configuraes do grfico 3D
ax.set_xlabel('X') # Configurando o rtulo do eixo x
ax.set_ylabel('Y') # Configurando o rtulo do eixo y
ax.set_zlabel('Z') # Configurando o rtulo do eixo z
ax.set_title('Cubo no Espao 3D') # Configurando o ttulo do grfico
ax.set_xlim(-0.5, 0.5) # Limites do eixo X
ax.set_ylim(-0.5, 0.5) # Limites do eixo Y
ax.set_zlim(-0.5, 0.5) # Limites do eixo Z
# Adicionando manualmente uma legenda para o eixo Z
ax.text(0.4, 0.4, 0.4, 'Z', color='black') # Adicionando o texto 'Z' na posio desejada
# Mostrando o grfico
plt.show()
```
