```python
import matplotlib.pyplot as plt
import numpy as np
#ALGORITMO DE RASTERIZAÇÃO DE CURVA DE BEZIER
#PONTOS DE CONTROLE
p0x = 0
p0y = 0
p1x = -1
p1y = 3
p2x = 2
p2y = 6
p3x = 4
p3y = 4
points_x = []
points_y = []


t = 0  # Valor inicial de t
step = 0.1  # Passo da iteração

while t <= 1:
    H_t = (2 * t**3 - 3 * t**2 + 1) * P1 + (t**3 - 2 * t**2 + t) * M1 + (-2 * t**3 + 3 * t**2) * P2 + (t**3 - t**2) * M2
    print(f"H({t:.2f}) = {H_t}")
    t += step  # Incremento de t
    points_x.append(P1)
    points_y.append(P2)

# Adiciona os pontos de controle ao gráfico
plt.scatter([p0x, p1x, p2x, p3x], [p0y, p1y, p2y, p3y], color='red', label='Pontos de Controle')
# Adiciona a curva de Bezier ao gráfico
plt.plot(points_x, points_y, label='Curva de Bezier')
# Configurações adicionais
plt.title('Curva de Bezier com Linha Conectando Pontos de Controle')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definição dos pontos e vetores tangentes
P1 = np.array([0, 0])
P2 = np.array([3, 3])
M1 = np.array([6, -6])
M2 = np.array([6, 4])

# Função da curva de Hermite
def hermite_curve(t, P1, P2, M1, M2):
    H1 = 2*t**3 - 3*t**2 + 1
    H2 = t**3 - 2*t**2 + t
    H3 = -2*t**3 + 3*t**2
    H4 = t**3 - t**2
    return H1*P1 + H2*M1 + H3*P2 + H4*M2

# Gerar pontos da curva
t_values = np.linspace(0, 1, 11)  # 10 segmentos (11 pontos)
curve_points = np.array([hermite_curve(t, P1, P2, M1, M2) for t in t_values])

# Plotar a curva
plt.plot(curve_points[:, 0], curve_points[:, 1], 'b-', label="Curva de Hermite")
plt.scatter([P1[0], P2[0]], [P1[1], P2[1]], color='red', label="Pontos P1 e P2")
plt.quiver(*P1, *M1, color='cyan', angles='xy', scale_units='xy', scale=1, label="M1")
plt.quiver(*P2, *M2, color='cyan', angles='xy', scale_units='xy', scale=1, label="M2")

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
