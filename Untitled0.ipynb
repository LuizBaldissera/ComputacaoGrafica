```python
import numpy as np
import matplotlib.pyplot as plt
# Função para calcular a translação dos pontos
def translacao(pontos, Tx, Ty):
  pontos_transladados = []
  for ponto in pontos:
    x_u = ponto[0]  Tx
    y_u = ponto[1] + Ty
    pontos_transladados.append((x_u, y_u))
  return pontos_transladados
# Pontos originais
p1 = (1, 3)
p2 = (3, 5)
p3 = (3, 1)
# Vetor de translação
Tx = 3
Ty = -4

# Calcular a translação dos pontos
pontos_transladados = translacao([p1, p2, p3, p1], Tx, Ty)
# Plotar os pontos originais e os pontos transladados
plt.plot([p1[0], p2[0], p3[0], p1[0]], [p1[1], p2[1], p3[1], p1[1]], 'bo-', label='Pontos originais')
plt.plot([ponto[0] for ponto in pontos_transladados], [ponto[1] for ponto in pontos_transladados],'ro-', label='Pontos transladados')
#[ponto[0] for ponto in pontos_transladados]:
#Isso é uma compreensão de lista, uma construção Python que permite criar listas de maneira mais
#compacta.
#O que está entre os colchetes é uma expressão que define como cada elemento da lista será
#construído.
#ponto[0] é a expressão usada para cada elemento da lista. Aqui, 'ponto' representa cada elemento
#da lista 'pontos_transladados', que são as coordenadas dos pontos transladados.
#ponto[0] extrai o primeiro elemento de cada tupla dentro da lista pontos_transladados, que
#corresponde à coordenada x do ponto.
#Então, [ponto[0] for ponto in pontos_transladados] cria uma lista contendo apenas as coordenadas
#x dos pontos transladados. Essa lista é usada para definir os valores do eixo x no gráfico.
#'bo-' especifica o estilo do gráfico: b para azul (blue), 'o' para marcador de círculo e '-' para linha
#contínua.
#'ro-':#Isso define o estilo do gráfico para os pontos transladados. Aqui, 'r' indica que os pontos
#serão vermelhos,
# 'o' especifica que os marcadores serão círculos e '-' indica que uma linha contínua conectará os
#pontos.
# Configurações do gráfico
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Translação de pontos no plano cartesiano')
plt.grid(True)
plt.legend()
# Mostrar o gráfico
plt.show()
```

```python

import matplotlib.pyplot as plt

def escala(pontos, Sx, Sy):
    pontos_escalados = []
    for ponto in pontos:
        x_u = ponto[0] * Sx
        y_u = ponto[1] * Sy
        pontos_escalados.append((x_u, y_u))
    return pontos_escalados

# Pontos originais (formando um quadrado)
p1 = (-2, -2)
p2 = (2, -2)
p3 = (2, 2)
p4 = (-2, 2)


Sx = -2
Sy = -2

# Calcular a escala dos pontos
pontos = [p1, p2, p3, p4]
pontos_escalados = escala(pontos, Sx, Sy)


quadrado_original = [p1, p2, p3, p4, p1]
quadrado_escalado = pontos_escalados + [pontos_escalados[0]]
plt.plot([ponto[0] for ponto in quadrado_original], [ponto[1] for ponto in quadrado_original], 'bo-', label='Quadrado original')
plt.plot([ponto[0] for ponto in quadrado_escalado], [ponto[1] for ponto in quadrado_escalado], 'ro-', label='Quadrado escalado')

# Configurações do gráfico
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Escala de quadrado no plano cartesiano')
plt.grid(True)
plt.legend()

# Mostrar o gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def rotacao(pontos, angulo):
    angulo_rad = np.radians(angulo)
    pontos_angulados = []

    for ponto in pontos:
        x, y = ponto  # Desempacotar os valores do ponto
        x_u = x * np.cos(angulo_rad) - y * np.sin(angulo_rad)
        y_u = y * np.cos(angulo_rad) + x * np.sin(angulo_rad)
        pontos_angulados.append((x_u, y_u))

    return pontos_angulados

# Pontos originais (formando um quadrado)
pontos = [(1, 1), (2, 2)]

# Solicita o ângulo de rotação ao usuário
angulo = int(input("Digite o ângulo de rotação: "))

# Aplica a rotação
pontos_rotacionados = rotacao(pontos, angulo)

# Plotando os pontos originais
plt.plot([p[0] for p in pontos], [p[1] for p in pontos], 'bo-', label='Original')

# Plotando os pontos rotacionados
plt.plot([p[0] for p in pontos_rotacionados], [p[1] for p in pontos_rotacionados], 'ro-', label='Rotacionado')

# Configurações do gráfico
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Rotação de {angulo}° no plano cartesiano')
plt.grid(True)
plt.legend()

# Mostrar o gráfico
plt.show()
```
